```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HR_Data.csv')
```

```python
# Eliminate NaN values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Identify and drop duplicate columns
df = df.loc[:,~df.columns.duplicated()]
df.columns
```

```python
one_uniq = dict()
for col in df.columns:
    one_uniq.update({col: df[col].unique().sum()})
        
print('The name of column(s) with value: %s' % one_uniq)
```

```python
one_uniq = dict()
for col in df.columns:
    if len(df[col].unique()) == 1:
        one_uniq.update({col: df[col].unique().tolist()})
        
print('The name of column(s) with single value: %s' % one_uniq)
```

```python
df.drop(one_uniq.keys(), axis=1, inplace=True)
```

```python
education_map = {1: 'Below College', 2: 'College',
                 3: 'Bachelor', 4: 'Master', 5: 'Phd'}
envir_sat_map = {1: 'Low', 2: 'Medium',
                 3: 'High', 4: 'Very High'}
job_inv_map = {1: 'Low', 2: 'Medium',
               3: 'High', 4: 'Very High'}
job_sat_map = {1: 'Low', 2: 'Medium',
               3: 'High', 4: 'Very High'}
perf_rate_map = {1: 'Low', 2: 'Good',
                 3: 'Excellent', 4: 'Outstanding'}
relation_sat_map = {1: 'Low', 2: 'Medium',
                    3: 'High', 4: 'Very High'}
work_life_map = {1: 'Bad', 2: 'Good',
                 3: 'Better', 4: 'Best'}

df['Education'] = df['Education'].apply(lambda x: education_map[x])

df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].apply(lambda x: envir_sat_map[x])

df['JobInvolvement'] = df['JobInvolvement'].apply(lambda x: job_inv_map[x])

df['JobSatisfaction'] = df['JobSatisfaction'].apply(lambda x: job_sat_map[x])

df['PerformanceRating'] = df['PerformanceRating'].apply(lambda x: perf_rate_map[x])

df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].apply(lambda x: relation_sat_map[x])

df['WorkLifeBalance'] = df['WorkLifeBalance'].apply(lambda x: work_life_map[x])
```

```python
#Univariate Analysis
le_10_val_col = [] #lees than
mt_10_val_col = [] # more than
obj_val_col = [] # object variable
for col in df.columns:
    if (len(df[col].unique()) <= 10) & (df[col].dtypes == 'int64'):
        le_10_val_col.append(col)
    elif (len(df[col].unique()) > 10) & (df[col].dtypes == 'int64'):
        mt_10_val_col.append(col)
    elif df[col].dtypes == 'O':
        obj_val_col.append(col)
        
```

```python
import numpy as np
from numpy import median
for i in mt_10_val_col:
    plt.figure(figsize=(10, 5))
    sns.distplot(df.loc[:, i], kde=True,color='y')
    plt.title(r'$\bf{%s}$ distribution with mean of %.2f and median of %.2f' % \
              (i, df.loc[:, i].mean(), median(df.loc[:, i])),
              size=14)
    plt.xlabel(i, fontsize=12)
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
    plt.axvline(median(df.loc[:, i]), 0, linestyle='--', 
                linewidth=1.5, color='r')
    plt.axvline(df.loc[:, i].mean(), 0, linestyle='--', 
                linewidth=1.5, color='g')
    plt.show()
```

```python
# IQR 
q1 = df['MonthlyIncome'].quantile(0.25)
q3 = df['MonthlyIncome'].quantile(0.75)
iqr = q3-q1
```

```python
q1, q3, iqr
```

```python
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)
lower_limit, upper_limit
```

```python
# capping - change the outlier values to upper (or) lower limit values
new_df = df.copy()
new_df.loc[(new_df['MonthlyIncome']>upper_limit), 'MonthlyIncome'] = upper_limit
new_df.loc[(new_df['MonthlyIncome']<lower_limit), 'MonthlyIncome'] = lower_limit
```

```python
sns.boxplot(new_df['MonthlyIncome'])
```
